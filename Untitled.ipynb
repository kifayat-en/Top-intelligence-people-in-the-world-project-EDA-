```python
gender_expertise = pd.crosstab(df['Gender'], df['Field of Expertise'])
gender_expertise = gender_expertise.reset_index().melt(id_vars='Gender', value_name='Count')


plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
ax = sns.barplot(data=gender_expertise, x='Gender', y='Count', hue='Field of Expertise', ci='sd')
```
